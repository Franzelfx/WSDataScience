## 6. Fehlende Werte erkennen

### Konzept: Missing Values (NaN)
In der Realität sind Datensätze selten vollständig. Fehlende Werte heißen **NaN** (Not a Number).  
Bevor wir analysieren, müssen wir wissen:
- **Wo** fehlen Werte?
- **Wie viele** fehlen?
- **Was tun wir damit?** (Löschen? Auffüllen?)

Das ist ein SEHR wichtiger Schritt – fehlende Werte können unsere Analysen verfälschen!

```python
# Fehlende Werte pro Spalte zählen
fehlende_werte = df.isnull().sum()
print("Fehlende Werte pro Spalte:")
print(fehlende_werte[fehlende_werte > 0])  # Nur Spalten mit fehlenden Werten anzeigen
```

### Umgang mit fehlenden Werten

Es gibt verschiedene Strategien:
1. **Löschen** – Zeilen oder Spalten entfernen (wenn wenige fehlen)
2. **Auffüllen (Imputation)** – z.B. mit dem Median oder Durchschnitt
3. **Kennzeichnen** – Eine eigene Kategorie "unbekannt" erstellen

Wir füllen das fehlende **Alter** mit dem **Median** auf – das ist robust gegen Ausreißer.

```python
# Fehlende Alterswerte mit dem Median auffüllen
median_alter = df["age"].median()
df["age"] = df["age"].fillna(median_alter)

print(f"Fehlende Alterswerte wurden mit dem Median ({median_alter}) aufgefüllt.")
print(f"Fehlende Werte in 'age' jetzt: {df['age'].isnull().sum()}")
```

## 10. Zusammenhänge erkennen – Korrelation

### Konzept: Korrelation
Die **Korrelation** misst, wie stark zwei Zahlen-Spalten zusammenhängen:
- **+1** = perfekter positiver Zusammenhang (wenn A steigt, steigt auch B)
- **0** = kein Zusammenhang
- **-1** = perfekter negativer Zusammenhang (wenn A steigt, sinkt B)

### Konzept: Heatmap
Eine **Heatmap** stellt die Korrelationsmatrix farblich dar – dunklere Farben = stärkerer Zusammenhang.

```python
# Korrelationsmatrix berechnen (nur numerische Spalten)
korrelation = df.select_dtypes(include=["number"]).corr()

# Heatmap erstellen
plt.figure(figsize=(10, 8))
sns.heatmap(korrelation, annot=True, fmt=".2f", cmap="RdBu_r", center=0, square=True)
plt.title("Korrelationsmatrix", fontsize=14)
plt.tight_layout()
plt.show()
```
