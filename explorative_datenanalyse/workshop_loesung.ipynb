```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")

print("✅ Alle Bibliotheken geladen!")
```

---
## 2. Daten laden

```python
df = sns.load_dataset("titanic")
print(f"Der Datensatz hat {df.shape[0]} Zeilen und {df.shape[1]} Spalten.")
```

---
## 3. Erste Eindrücke – Daten anschauen

```python
# ✅ LÖSUNG: Letzte 5 Zeilen anzeigen
df.tail()
```

---
## 4. Datentypen und Struktur verstehen

```python
df.info()
```

```python
# ✅ LÖSUNG: Datentypen aller Spalten
df.dtypes
```

---
## 5. Deskriptive Statistik

```python
df.describe()
```

```python
# ✅ LÖSUNG: Durchschnittlicher Ticketpreis
durchschnittspreis = df["fare"].mean()
print(f"Der durchschnittliche Ticketpreis betrug: {durchschnittspreis:.2f}")
```

```python
# ✅ LÖSUNG: Medianalter
# Antwort auf die Frage: Der Median ist robuster gegen Ausreißer.
# Wenn z.B. ein Passagier 200 Jahre alt wäre (Datenfehler), würde der
# Durchschnitt stark beeinflusst, der Median aber kaum.
medianalter = df["age"].median()
print(f"Das Medianalter der Passagiere war: {medianalter:.1f} Jahre")
```

---
## 6. Fehlende Werte erkennen

```python
fehlende_werte = df.isnull().sum()
print("Fehlende Werte pro Spalte:")
print(fehlende_werte[fehlende_werte > 0])
```

```python
# ✅ LÖSUNG: Prozentuale fehlende Werte
fehlende_prozent = (df.isnull().sum() / len(df)) * 100
print("Fehlende Werte in Prozent:")
print(fehlende_prozent[fehlende_prozent > 0].round(1))
```

```python
# Fehlende Alterswerte mit dem Median auffüllen
median_alter = df["age"].median()
df["age"] = df["age"].fillna(median_alter)

print(f"Fehlende Alterswerte wurden mit dem Median ({median_alter}) aufgefüllt.")
print(f"Fehlende Werte in 'age' jetzt: {df['age'].isnull().sum()}")
```

---
## 7. Visualisierung – Verteilungen

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="age", bins=30, kde=True, color="steelblue")
plt.title("Altersverteilung der Titanic-Passagiere", fontsize=14)
plt.xlabel("Alter (Jahre)")
plt.ylabel("Anzahl Passagiere")
plt.show()
```

```python
# ✅ LÖSUNG: Histogramm für Ticketpreise
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="fare", bins=30, kde=True, color="steelblue")
plt.title("Verteilung der Ticketpreise", fontsize=14)
plt.xlabel("Ticketpreis")
plt.ylabel("Anzahl Passagiere")
plt.show()
```

---
## 8. Visualisierung – Kategorien vergleichen

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="class", order=["First", "Second", "Third"], palette="viridis")
plt.title("Anzahl Passagiere pro Klasse", fontsize=14)
plt.xlabel("Passagierklasse")
plt.ylabel("Anzahl")
plt.show()
```

```python
# ✅ LÖSUNG: Überlebensrate nach Geschlecht
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="sex", y="survived", palette="Set2")
plt.title("Überlebensrate nach Geschlecht", fontsize=14)
plt.xlabel("Geschlecht")
plt.ylabel("Überlebensrate")
plt.show()
```

---
## 9. Visualisierung – Boxplot

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="class", y="fare", order=["First", "Second", "Third"], palette="coolwarm")
plt.title("Ticketpreise nach Passagierklasse", fontsize=14)
plt.xlabel("Passagierklasse")
plt.ylabel("Ticketpreis")
plt.show()
```

```python
# ✅ LÖSUNG: Boxplot Alter nach Überlebensstatus
# Erkenntnis: Der Altersunterschied ist gering, aber jüngere Passagiere
# (besonders Kinder) hatten tendenziell bessere Überlebenschancen.
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="survived", y="age", palette="Set2")
plt.title("Altersverteilung nach Überlebensstatus", fontsize=14)
plt.xlabel("Überlebt (0=Nein, 1=Ja)")
plt.ylabel("Alter (Jahre)")
plt.show()
```

---
## 10. Zusammenhänge erkennen – Korrelation

```python
korrelation = df.select_dtypes(include=["number"]).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(korrelation, annot=True, fmt=".2f", cmap="RdBu_r", center=0, square=True)
plt.title("Korrelationsmatrix", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# ✅ LÖSUNG: Stärkste Korrelation mit "survived"
# Ergebnis: "fare" hat die stärkste positive Korrelation (höherer Preis → bessere Überlebenschance),
# "pclass" die stärkste negative (niedrigere Klasse = höhere Zahl → schlechtere Chance)
korrelation["survived"].sort_values(ascending=False)
```

---
## 11. Gruppenvergleiche

```python
ueberlebensrate_klasse = df.groupby("class")["survived"].mean() * 100
print("Überlebensrate nach Klasse (in %):")
print(ueberlebensrate_klasse.round(1))
```

```python
# ✅ LÖSUNG: Überlebensrate nach Geschlecht
ueberlebensrate_geschlecht = df.groupby("sex")["survived"].mean() * 100
print("Überlebensrate nach Geschlecht (in %):")
print(ueberlebensrate_geschlecht.round(1))
```

```python
# ✅ LÖSUNG: Überlebensrate nach Klasse UND Geschlecht
ueberlebensrate_kombi = df.groupby(["class", "sex"])["survived"].mean() * 100
print("Überlebensrate nach Klasse und Geschlecht (in %):")
print(ueberlebensrate_kombi.round(1))
```

---
## 12. Kombinierte Visualisierung

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="class", y="survived", hue="sex",
            order=["First", "Second", "Third"], palette="Set2")
plt.title("Überlebensrate nach Klasse und Geschlecht", fontsize=14)
plt.xlabel("Passagierklasse")
plt.ylabel("Überlebensrate")
plt.legend(title="Geschlecht")
plt.show()
```

```python
# ✅ LÖSUNG: Altersverteilung nach Überlebensstatus
# Erkenntnis: Bei den jüngsten Passagieren (Kinder) ist ein deutlicher
# Überlebensvorteil sichtbar → "Frauen und Kinder zuerst"
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="age", hue="survived", bins=30, kde=True)
plt.title("Altersverteilung nach Überlebensstatus", fontsize=14)
plt.xlabel("Alter (Jahre)")
plt.ylabel("Anzahl Passagiere")
plt.show()
```

---
## 13. Zusammenfassung und Erkenntnisse

```python
# ✅ LÖSUNG: Gesamtüberlebensrate
gesamt_rate = df["survived"].mean() * 100
print(f"Gesamtüberlebensrate: {gesamt_rate:.1f}%")
```

### Musterlösung – Erkenntnisse

1. **Anzahl Passagiere:** 891
2. **Fehlende Werte:** `age` (~19.9%), `deck` (~77.2%), `embarked` (~0.2%), `embark_town` (~0.2%)
3. **Gesamtüberlebensrate:** ~38.4%
4. **Höchste Überlebensrate:** Frauen in der 1. Klasse (~96.8%)
5. **Stärkste Korrelation:** `fare` (positiv) und `pclass` (negativ)
6. **Überraschend:** Der enorme Unterschied zwischen Klassen und Geschlechtern – "Frauen und Kinder zuerst" zeigt sich klar in den Daten

## Bonus-Aufgaben – Lösungen 🌟

```python
# ✅ BONUS 1: Altersgruppen erstellen und Überlebensrate visualisieren
df["age_group"] = pd.cut(df["age"], bins=[0, 12, 18, 35, 60, 100],
                          labels=["Kind", "Jugendlich", "Jung-Erwachsen", "Erwachsen", "Senior"])

plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="age_group", y="survived", palette="viridis")
plt.title("Überlebensrate nach Altersgruppe", fontsize=14)
plt.xlabel("Altersgruppe")
plt.ylabel("Überlebensrate")
plt.show()

# Erkenntnis: Kinder hatten die höchste Überlebensrate
```

```python
# ✅ BONUS 2: Pairplot
sns.pairplot(df[["age", "fare", "survived"]], hue="survived", palette="Set2")
plt.suptitle("Pairplot: Alter, Ticketpreis und Überleben", y=1.02, fontsize=14)
plt.show()

# Erkenntnis: Passagiere mit höheren Ticketpreisen überlebten häufiger
```
